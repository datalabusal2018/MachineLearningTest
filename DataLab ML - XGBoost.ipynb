```python
import pandas as pd
import numpy as np

import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
```

```python
dataTrain=pd.read_csv("/Users/JuanCB/GitHub/MachineLearningTest/data/train.csv", low_memory=False)
dataTest=pd.read_csv("/Users/JuanCB/GitHub/MachineLearningTest/data/test.csv", low_memory=False)
```

```python
dataTrain.isnull()
```

```python
dataTrain = dataTrain.fillna(dataTrain.mean())

dataTest = dataTest.replace('?', np.nan)
dataTest = dataTest.fillna(dataTest.mean())
#dataTest = dataTest.dropna()
```

```python
display(dataTest)
```

```python
list_features = list(dataTrain)
variable2predict = 'num'

list_features.remove(variable2predict)
#display(columnas)

features_gen = pd.Index(list_features)
display(features_gen)
```

```python
dataTotal = dataTrain[features_gen].append(dataTest[features_gen])

# In the following slicing, we take rows from 0 to the Nth row 
# where N is the number of rows in dataTrain (.shape[0] gets the rows number)
train_X = dataTotal[0:dataTrain.shape[0]].as_matrix()
# In the following slicing, we take rows from from Nth row to the dataframe's final row
# where N is the number of rows in dataTrain (.shape[0] gets the rows number)
test_X = dataTotal[dataTrain.shape[0]::].as_matrix()
train_y = dataTrain['num']
```

```python
model = xgb.XGBClassifier(max_depth=3, n_estimators=300, learning_rate=0.05).fit(train_X, train_y)
predictions = model.predict(test_X)
```

```python
num_predict = pd.DataFrame({ 'Pacient ID': dataTest.index.tolist(),
                             'Predicted-Num': predictions })
display(num_predict)
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import StratifiedKFold
from sklearn.cross_validation import cross_val_score
from sklearn.metrics import auc

parameters = {
    'max_depth': [1, 2, 3],
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.1, 0.5],
}

clf = GridSearchCV(model, parameters, n_jobs=1, cv=2)
    
clf.fit(train_X, train_y)
best_parameters, score, _ = max(clf.grid_scores_, key=lambda x: x[1])
print('score:', score)
for param_name in sorted(best_parameters.keys()):
    print("%s: %r" % (param_name, best_parameters[param_name]))

    
print('predicted:', clf.predict(test_X))
```
